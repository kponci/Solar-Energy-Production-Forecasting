```python
# H2O

all_predictions = pd.DataFrame()
# Init of H2O
h2o.init(max_mem_size = "16G") # If u have more RAM change the parameter

# Load data
print(f"Training on dataset")
# Load data from Parquet files and concatenate them into a single DataFrame 'X'
# X = pd.concat([
#     pd.read_parquet(f"../../dataset/{letter}/X_train_observed.parquet"),
#     pd.read_parquet(f"../../dataset/{letter}/X_train_estimated.parquet")
# ], ignore_index=True)
# y = pd.read_parquet(f"../../dataset/{letter}/train_targets.parquet")
def mean_15minX(X):
    ret = X.drop('date_calc', axis=1, errors='ignore').iloc[:,1:].groupby(np.arange(len(X))//4).mean()
    ret.insert(0, 'date_forecast',list(X.loc[::4, 'date_forecast']))
    return ret


XA1 = mean_15minX(pd.read_parquet(f"../../dataset/A/X_train_observed.parquet"))
XA1['locationA'] = True
XA1['locationB'] = False
XA1['locationC'] = False
XA1['observed'] = True
XB1 = mean_15minX(pd.read_parquet(f"../../dataset/B/X_train_observed.parquet"))
XB1['locationA'] = False
XB1['locationB'] = True
XB1['locationC'] = False
XB1['observed'] = True
XC1 = mean_15minX(pd.read_parquet(f"../../dataset/C/X_train_observed.parquet"))
XC1['locationA'] = False
XC1['locationB'] = False
XC1['locationC'] = True
XC1['observed'] = True
XA2 = mean_15minX(pd.read_parquet(f"../../dataset/A/X_train_estimated.parquet"))
XA2['locationA'] = True
XA2['locationB'] = False
XA2['locationC'] = False
XA2['observed'] = False
XB2 = mean_15minX(pd.read_parquet(f"../../dataset/B/X_train_estimated.parquet"))
XB2['locationA'] = False
XB2['locationB'] = True
XB2['locationC'] = False
XB2['observed'] = False
XC2 = mean_15minX(pd.read_parquet(f"../../dataset/C/X_train_estimated.parquet"))
XC2['locationA'] = False
XC2['locationB'] = False
XC2['locationC'] = True
XC2['observed'] = False
XAT = mean_15minX(pd.read_parquet(f"../../dataset/A/X_test_estimated.parquet"))
XAT['locationA'] = True
XAT['locationB'] = False
XAT['locationC'] = False
XAT['observed'] = False
XBT = mean_15minX(pd.read_parquet(f"../../dataset/B/X_test_estimated.parquet"))
XBT['locationA'] = False
XBT['locationB'] = True
XBT['locationC'] = False
XBT['observed'] = False
XCT = mean_15minX(pd.read_parquet(f"../../dataset/C/X_test_estimated.parquet"))
XCT['locationA'] = False
XCT['locationB'] = False
XCT['locationC'] = True
XCT['observed'] = False
yA = pd.read_parquet(f"../../dataset/A/train_targets.parquet")
yB = pd.read_parquet(f"../../dataset/B/train_targets.parquet")
yC = pd.read_parquet(f"../../dataset/C/train_targets.parquet")



XA = pd.concat([
    XA1, XA2
], ignore_index=True)
XB = pd.concat([
    XB1, XB2
], ignore_index=True)
XC = pd.concat([
    XC1, XC2
], ignore_index=True)
# Preprocess the data


# Merge DataFrame 'X' and 'y' based on the common column "date_forecast"
XYA = XA.merge(yA, left_on = "date_forecast", right_on = "time")
XYB = XA.merge(yB, left_on = "date_forecast", right_on = "time")
XYC = XA.merge(yC, left_on = "date_forecast", right_on = "time")
X_y_bacon = pd.concat([XYA,XYB, XYC])
X_test = pd.concat([XAT, XBT, XCT])
x_test = h2o.H2OFrame(X_test)


# Split data into training and validation frames (80:20)
train,test = sklearn.model_selection.train_test_split(X_y_bacon, test_size = 0.20, random_state=42)

train_frame = h2o.H2OFrame(train)
test_frame = h2o.H2OFrame(test)

x = train_frame.columns[:-1] 
y = train_frame.columns[-1] 

# Create an AutoML model
aml = H2OAutoML(max_runtime_secs = 600,
                sort_metric = "MAE",
                stopping_metric = "MAE",
                seed = 7213712285) 
aml.train(x = x, 
          y = y,
          training_frame = train_frame,
          validation_frame = test_frame)

print(aml.leaderboard)
best_model = aml.get_best_model(criterion='MAE')

# Make predictions on test data and make CSV file
prediction1 = best_model.predict(x_test)
predictions1_df = h2o.as_list(prediction1)
predictions1_df[predictions1_df < 0.] = 0.

all_predictions = pd.concat([all_predictions, predictions1_df], ignore_index=True)
all_predictions.to_csv('alltogetherh2o.csv', index_label='id')
print("CSV file updated")

# Explain the best model on the validation frame for model interpretation
best_model.explain(test_frame)

# Shut down H2O
h2o.shutdown()
print("Done")
```
